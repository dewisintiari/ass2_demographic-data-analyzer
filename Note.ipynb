### The data set

```python
df = pd.read_csv("adult.data.csv")
print(df)
```

### How many people of each race are represented in this dataset? This should be a Pandas series with race names as the index labels. (`race` column)

```python
print(df['race'])
```

```python
race_count = df['race'].value_counts()
display(race_count)
```

### What is the average age of men?

```python
df.loc[df['sex'] == 'Male', 'age'].mean()
```

```python
num = df.loc[df['sex'] == 'Male', 'age'].count()
average_age_men = num/len(df.index)
print(average_age_men)
```

### What is the percentage of people who have a Bachelor's degree?

```python
bachelor_df = df.loc[df['education'] == 'Bachelors']
percentage_bachelors = len(bachelor_df.index) / len(df.index)

print(percentage_bachelors)
```

### What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

### What percentage of people without advanced education make more than 50K?

```python
higher_df = df.loc[(df['education'] == 'Bachelors') | (df['education'] == 'Masters') | (df['education'] == 'Doctorate')]
display(higher_df)
```

```python
lower_df = df.loc[(df['education'] != 'Bachelors') & (df['education'] != 'Masters') & (df['education'] != 'Doctorate')]
display(lower_df)
```

```python
higher_education_rich = higher_df.loc[df['salary'] == '>50K', 'education'].count()  / len(df.index) * 100
print(higher_education_rich)
```

```python
lower_education_rich = lower_df.loc[df['salary'] == '>50K', 'education'].count() / len(df.index) * 100
print(lower_education_rich)
```

### What is the minimum number of hours a person works per week?

```python
min_work_hours = df['hours-per-week'].min()
```

### What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
# dataframe of people who work in minimum hours, but get high salary >50K

minWorkHours_df = df.loc[(df['hours-per-week'] == min_work_hours) & (df['salary'] == '>50K')]

display(minWorkHours_df)
```

```python
# dataframe of people who work in minimum hours, but get high salary >50K

minWorkHours_df = df.loc[df['hours-per-week'] == min_work_hours]
rich_df = minWorkHours_df.loc[minWorkHours_df['salary'] == '>50K']

display(rich_df)
```

```python
num_min_workers = len(rich_df.index) / len(df.index)

rich_percentage = len(rich_df.index) / len(minWorkHours_df.index) * 100

print(num_min_workers, 'and', rich_percentage)
```

### What country has the highest percentage of people that earn >50K and what is that percentage?

```python
richCountry_df = df.loc[df['salary'] == '>50K']['native-country']

#to create new dataframe with ne indexes names
richRank_df = richCountry_df.value_counts().rename_axis('Country').to_frame('Counts')

display(richCountry_df)
display(richRank_df)
```

```python
highest_earning_country = richRank_df['Counts'].idxmax()
print(highest_earning_country)
```

```python
salary_df = df[['native-country', 'salary']]
display(salary_df)
```

#### Below is the real code to answer the questions

```python
recapSalary_df = df['native-country'].fillna(0).value_counts()
display(recapSalary_df)
```

```python
highSalary_df = df[df['salary'] == '>50K']['native-country'].fillna(0).value_counts()
display(highSalary_df)
```

```python
highest_earning_country = highSalary_df.idxmax()
print(highest_earning_country)
```

```python
highestEarningCountryPercentage_df = (100*highSalary_df) / (recapSalary_df)
highest_earning_country = highestEarningCountryPercentage_df.idxmax()
highest_earning_country_percentage = highestEarningCountryPercentage_df[highest_earning_country]

display(highestEarningCountryPercentage_df)
print(highest_earning_country)
print(highest_earning_country_percentage)
```

### Identify the most popular occupation for those who earn >50K in India.

```python
richIndiaOccupation_df = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')]['occupation']
display(richIndiaOccupation_df)
```

```python
recapRichIndiaOccup_df = richIndiaOccupation_df.value_counts()
display(recapRichIndiaOccup_df)

top_IN_occupation = recapRichIndiaOccup_df.idxmax()
print(top_IN_occupation)
```
